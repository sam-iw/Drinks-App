```python
age_data = %sql select drinker_age from people;

labels = '0-10', '11 - 17', '18 - 30', '31+'
kids = 0
teens = 0
young_adults = 0
adults = 0
for row in age_data:
    if row[0] < 11:
        kids += 1
    elif row [0] < 18:
        teens += 1
    elif row [0] < 31:
        young_adults += 1
    else:
        adults += 1
sizes = [kids, teens, young_adults, adults]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
print(int(kids))
print(int(adults))
plt.figure(figsize=(10,10))
plt.show()
```
